```python
import pandas as pd
import matplotlib.pyplot as plt

try: 
    data = pd.read_csv('Farm_data.csv') 
except FileNotFoundError:
    print("File Not Found")
```

```python
print(data.head())

print(data.info()) 

print("Missing values: \n", data.isnull().sum)

data_cleaned = data.dropna() 
```

```python
data.describe()
```

```python
grouped_mean = data.groupby('Type of crop').mean(numeric_only=True)
print(grouped_mean)

grouped_sum = data.groupby('Type of crop').sum(numeric_only=True)
print(grouped_sum)
```

```python
grouped_sum.sort_values('Average yield (kg/ha)', ascending=True)
```

```python
plt.bar(data['FarmNumber'], data['Production (tonnes)'], color='Purple') 
plt.title('Production per farm')
plt.xlabel("Farm number")
plt.ylabel("Production(tonnes)")
plt.show()
```

```python
plt.plot(data['FarmNumber'], data['Seeded area (hectares)'], marker = 'o')
plt.title('Seeded area per farm')
plt.xlabel('Farm Number')
plt.ylabel('Seeded area(ha)')
```

```python
plt.pie(data['Average yield (kg/ha)'], labels=data['FarmNumber'])
plt.title("Average yield for each farm")
plt.show()
```

```python
plt.hist(data['Average yield (kg/ha)'], bins=11, color='cyan')
plt.title("Average yield distribution")
plt.xlabel("Average yield(kg/ha)")
plt.ylabel("Frequency")
plt.show()
```
